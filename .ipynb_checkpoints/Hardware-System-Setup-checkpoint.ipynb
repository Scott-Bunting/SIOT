### Data Collection: *01:30am 5th December*

At this time there was a large difference between indoor and outdoor air temperature. This resulted in a large, and fast, drop in temperature. Initially I thought this was a natural response, but upon further investigation it became clear that the response is as shown due to the sensor being mounted to the Raspberry Pi. 
 
 
This means that the rapid cooling is as a result of the cold draft, from the opening of the window, convection cooling the system. Therefore, as soon as the convection current is removed (window closed) the system quickly returns to its previous temperature. The sensor is heated both convectively and via conductance, from the main body of the Raspberry Pi.

```python
plt.figure(1)
plt.plot_date(datetime[30:997],temp[30:997])
plt.title('Temperature C')
plt.figure(2)
plt.plot_date(datetime[30:997],hum[30:997])
plt.title('Humidity %RH')
plt.figure(3)
plt.plot_date(datetime[30:997],pres[30:997])
plt.title('Pressure hPA')
```

### Data Collection: *11am 5th December*

Repeated same experiment as above during the day, when the temperature differential between inside and outside is lower. Returns similar results but with a smaller change in temperature.

Fluctuations in temperature believed to be a result of computational power fluctuations, as I was running a separate terminal to install packages on the device.
